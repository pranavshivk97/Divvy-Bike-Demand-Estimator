```python
import pandas as pd
import numpy as np
pd.options.mode.chained_assignment = None
```

# Dataset 1: Divvy Bike Data

```python
# Read the various csv files from Nov 2019 to Oct 2020 (includes Oct 2019 as well)
d1 = pd.read_csv("Divvy_Trips_2019_Q4.csv")
d2 = pd.read_csv("divvydata_Jan-Mar2020.csv")
d3 = pd.read_csv("202004-divvy-tripdata.csv")
d4 = pd.read_csv("202005-divvy-tripdata.csv")
d5 = pd.read_csv("202006-divvy-tripdata.csv")
d6 = pd.read_csv("202007-divvy-tripdata.csv")
d7 = pd.read_csv("202008-divvy-tripdata.csv")
d8 = pd.read_csv("202009-divvy-tripdata.csv")
d9 = pd.read_csv("202010-divvy-tripdata.csv")
```

```python
# d1 (2019 Quarter data) has some columns with differin names; renaming to match the other data
d1.rename({'start_station_nam': 'start_station_name'}, inplace=True, axis=1)
d1 = d1[['ride_id', 'started_at', 'ended_at', 'start_station_name', 'start_station_id', 'end_station_name', 'end_station_id', 'member_casual']]
d1.head()
```

```python
# merge all the datasets for the months to get the total data for the Divvy bikes
divvy_data = pd.concat([d1, d2, d3, d4, d5, d6, d7, d8, d9])
divvy_data
```

# Dataset 2 - Weather Data

```python
# read the weather data from the csv file into a pandas dataframe
weather = pd.read_csv("weather.csv")
weather
```

```python
weather = weather[['Date', 'Temperature', 'Humidity', 'Wind Speed', 'Wind Gust', 'Precipitation', 'Condition']]
weather.head()
```

```python
# Convert the date columns in the divvy data to datetime format 
divvy_data['started_at'] = pd.to_datetime(divvy_data.started_at)
divvy_data['ended_at'] = pd.to_datetime(divvy_data.ended_at)
divvy_data['Year'] = divvy_data.started_at.dt.year
divvy_data['Month'] = divvy_data.started_at.dt.month
divvy_data['Day'] = divvy_data.started_at.dt.day
divvy_data['Hour'] = divvy_data.started_at.dt.hour

# save the dataframe to a csv
divvy_data.to_csv("divvy_final.csv", index=False)
divvy_data
```

```python
# convert the weather date to datetime format
weather['Date'] = pd.to_datetime(weather.Date)
weather['Year'] = weather.Date.dt.year
weather['Month'] = weather.Date.dt.month
weather['Day'] = weather.Date.dt.day
weather['Hour'] = weather.Date.dt.hour

# save the dataframe to a csv file
weather.to_csv("weather_final.csv", index=False)
weather
```

```python
# merge both the divvy and weather data based on the extracted year, month, day and hour
merged_df = divvy_data.merge(weather, how='inner', on=['Year', 'Month', 'Day', 'Hour'])
merged_df.drop(['Date'], axis=1, inplace=True)
merged_df
```

# Dataset 3: Holidays

```python
hols = pd.read_csv('holidays.csv')
hols
```

```python
hols.Date = pd.to_datetime(hols.Date)
hols['Year'] = hols.Date.dt.year
hols['Month'] = hols.Date.dt.month
hols['Day'] = hols.Date.dt.day
hols
```

```python
final_merged_data = merged_df.merge(hols, how='inner', on=['Year', 'Month', 'Day'])
final_merged_data.drop("Date", axis=1, inplace=True)
final_merged_data
```

```python
final_merged_data.to_csv("final_merged_data.csv", index=False)
```
